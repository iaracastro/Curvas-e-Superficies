```python
def get_offset_curve(alpha, d):
    d_alpha = diff(alpha, t)

    T = d_alpha/d_alpha.norm()
    N = diff(T, t)/diff(T, t).norm()
    offset = alpha + d*N
    return offset
```

```python
def plot_multiple_d(alpha, d, k, intervalo, size):   
    figures = []
    curve = parametric_plot(alpha, (t, intervalo[0], intervalo[1]), figsize=size, color="blue")
    d_0 = d
    for i in range(1,k+1):
        d = d*i
        positive_offset = get_offset_curve(alpha, d)
        negative_offset = get_offset_curve(alpha, -d)
        
        figures.append(parametric_plot((positive_offset), (t, intervalo[0], intervalo[1]), figsize=size, color="green"))   
        figures.append(parametric_plot((negative_offset), (t, intervalo[0], intervalo[1]), figsize=size, rcolor="red")) 
        d = d_0
    return curve + sum(figures)
```

### Example

```python
alpha(t) = (t,t^2)
intervalo = [-2,2]
k = 2
d = 0.5

plot_multiple_d(alpha, d, k, intervalo,7)
```
